# Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.metrics import r2_score 
```

```python
# Import 'all_data.csv'
df = pd.read_csv('all_data.csv')
```

```python
# EDA
print(df.head())
print(df.describe())
```

```python
# Rename Life Expectancy column
dict = {'Country': 'country',
        'Year': 'year',
        'Life expectancy at birth (years)': 'life_expectancy'}

df.rename(columns = dict, inplace = True)

# print top by rows
print(df.head())
```

## Analysis 1: Has life expectancy increased over time in the six nations?

```python
# Create variables
countries = df.country.unique()
years = df.year.unique()
color = ['b', 'g', 'r', 'c', 'm', 'y']
print(countries)
print(years)
```

```python
# Create for loop variables
rows = 2
columns = 3
index = 1

plt.figure(figsize=(20, 16))

# Loop through countries to create plot
for i in countries:
    plt.subplot(rows, columns, index)
    plt.plot(years, df.life_expectancy[df.country == countries[index-1]], color = 'white', marker = 'o', markerfacecolor = color[index-1])
    plt.title(f'Life Expectancy in {countries[index-1]} \n{years[0]} - {years[-1]}')
    plt.xlabel('Year')
    plt.ylabel('Life Expectancy (Years)')
    index += 1

plt.show()
plt.clf()
```

## Analysis 2: Has GDP increased over time in the six nations?

```python
# Create for loop variables
rows = 2
columns = 3
index = 1

plt.figure(figsize = (20, 16))

# Loop through countries to create plot
for i in countries:

    # Create country dataframe
    country_df = df[df.country == countries[index-1]]
    
    # Tranform GDP into log
    log_gdp = country_df.GDP[country_df.GDP > 0]
    log_gdp = np.log(log_gdp)
    
    #Create plot
    plt.subplot(rows, columns, index)
    plt.plot(years, log_gdp, color = 'white', marker = 'o', markerfacecolor = color[index-1])
    plt.title(f'GDP in {countries[index-1]} \n{years[0]} - {years[-1]}')
    plt.xlabel('Year')
    plt.ylabel('GDP (logarithmic)')
    index += 1
    
plt.show()
plt.clf()
```

## Analysis 3: Is there a correlation between GDP and life expectancy of a country?

```python
# Create for loop variables
rows = 2
columns = 3
index = 1

plt.figure(figsize = (20, 16))

for i in countries:
    # Create country dataframe
    country_df = df[df.country == countries[index-1]]

    # Tranform GDP into log
    log_gdp = country_df.GDP[country_df.GDP > 0]
    log_gdp = np.log(log_gdp)

    # Calculate Linear Regression
    x = log_gdp
    y = country_df.life_expectancy
    
    #Calculate Slope & Intercept
    m, b = np.polyfit(x, y, 1)
        
    #Calculate R-squared
    regression_calc = m*x+b
    r2 = round(r2_score(y, regression_calc),2)
    
    regression_line_str = str(round(m, 2)) + 'x + ' + str(round(b, 2))

    # Create plot
    plt.subplot(rows, columns, index)
    plt.scatter(x, y, color = color[index-1])
    plt.plot(x, regression_calc, color = 'black')
    plt.title(f'{countries[index-1]} - GDP vs. Life Expectancy \ny = {regression_line_str} \nR-squared = {r2}')
    plt.xlabel('GDP (logarithmic)')
    plt.ylabel('Life Expectancy (years)')
    index += 1

plt.show()
plt.clf()
```

## Analysis 4: What is the average life expectancy in these nations?

```python
# Create for loop variables
rows = 2
columns = 3
index = 1

# Create for loop
for i in countries:
    
    # Create country dataframe
    country_df = df[df.country == countries[index-1]]

    #Calculate average life expectancy
    avg_life_expectancy = int(country_df.life_expectancy.mean())
    
    #Print life expectancy statement
    print(f'The average life expectancy in {countries[index-1]} is {avg_life_expectancy} years old.')
    
    index += 1
```

## Analysis 5: What is the distribution of that life expectancy?

```python
# Create for loop variables
rows = 2
columns = 3
index = 1

plt.figure(figsize = (20, 16))

# Create for loop
for i in countries:
    
    # Create country dataframe
    country_df = df[df.country == countries[index-1]]

    #Calculate average life expectancy
    life_expectancy = country_df.life_expectancy
    avg_life_expectancy = round(life_expectancy.mean(), 2)

    # Plot Life Expectancy
    ax = plt.subplot(rows, columns, index)
    plt.hist(life_expectancy, color = color[index-1])
    plt.axvline(avg_life_expectancy, color = 'black', linestyle = 'dashed')
    plt.title(f'{countries[index-1]} - Distribution of Life Expectancy \nMean Life Expectancy: {avg_life_expectancy}' )
    plt.xlabel('Life Expectancy (years)')
    index += 1
    
plt.show()
plt.clf()
```
